```r
library(raster)
library(ggplot2)
library(ggmap)
library(rgdal)
library(dplyr)
library(sf)
```

```r
roads = st_read(dsn = "tl_2016_us_primaryroads")
parks = st_read(dsn = "nps_boundary_shp/")
cntys = st_read(dsn = "~/Downloads/tl_2016_us_county")
```

```r
parks = st_transform(parks, st_crs(cntys)) %>% filter(UNIT_TYPE == "National Park")
parks = parks %>% mutate(parkarea = st_area(geometry))
joined = st_join(parks, cntys, join = st_overlaps)

        
```

```r
joined = joined %>% group_by(NAMELSAD) %>% summarise(area_mean = mean(parkarea))
```

```r
ggplot(parks %>% filter(parkarea > 1355398919)) +
    geom_sf(aes(fill='red')) +
    scale_x_continuous(limits = c(-120, -60)) +
    scale_y_continuous(limits = c(20, 70)) 
```
